```python
import pandas as pd

df = pd.read_csv('us-states.csv')
df
```

```python
df_date = df['date'].str.split('-',expand = True)

df_date.columns = ['year','month','day']

 

df_concat = pd.concat([df,df_date],axis =1)

df_concat['year'] = df_concat['year'].astype(int)

df_concat['month'] = df_concat['month'].astype(int)

df_concat['day'] = df_concat['day'].astype(int)

 

df_concat['date'] = pd.to_datetime(df_concat['date'])

df_cases = df_concat.sort_values(by='date')

df_cases
```

```python


 

# takes the max of all columns after groupby

df_cases = df_concat.groupby(['state','year','month']).max().reset_index()

 

df_dict  = {}

for state in df_cases['state'].unique():

    df_dict[state] = df_cases[df_cases['state']==state]

 

for state in df_dict.keys():

    df_dict[state]['Lag'] = df_dict[state]['cases'].shift(1).fillna(0)

    df_dict[state]['new_cases'] = df_dict[state]['cases'] - df_dict[state]['Lag']

 

df_final = pd.concat(df_dict.values())
```
